Esercizio Il dataset insurance.csv contiene dati rispetto a caratteristiche e abitudini delle persone, e della zona in cui vivono, rispetto ai costi individuali per le cure mediche come premio per le assicurazioni sulla salute. • Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; • Quali sono le medie di charges rispetto a region? Ci sono differenze significative? • E rispetto a smoker? E a sex? • Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?

```python
import pandas as pd

df = pd.read_csv("insurance.csv")

# Dimensioni
print("Dimensioni:", df.shape)

# Anteprima
print(df.head())

# Nomi colonne
print(df.columns.tolist())

```

```python
#Medie di charges rispetto a region
df.groupby("region")["charges"].mean()
```

Le regioni southeast e southwest tendono ad avere costi più alti, mentre northeast e northwest leggermente più bassi.

```python
#Medie di charges rispetto a smoker

df.groupby("smoker")["charges"].mean()
```

smoker = yes → costi molto più alti

smoker = no → costi molto più bassi

```python
#Medie di charges rispetto a sex

df.groupby("sex")["charges"].mean()
```

Differenze piccole: uomini e donne hanno costi simili.

```python
#Descrittori statistici di bmi

df["bmi"].describe()
```

```python
#Quartili del BMI

quartili = df["bmi"].quantile([0.25, 0.50, 0.75])
print(quartili)
```

```python
#Minimo, media e massimo di charges per quartile di BMI
#Prima creiamo una colonna che assegna ogni persona al suo quartile:

df["bmi_quartile"] = pd.qcut(df["bmi"], 4, labels=["Q1", "Q2", "Q3", "Q4"])
#Poi calcoliamo i descrittori:
df.groupby("bmi_quartile")["charges"].agg(["min", "mean", "max"])
```

```python
df["bmi_quartile"] = pd.qcut(df["bmi"], 4, labels=["Q1", "Q2", "Q3", "Q4"])
df.groupby("bmi_quartile", observed=True)["charges"].agg(["min", "mean", "max"])
```

Esercizio 1/2 Il dataset pokemon.csv contiene un database di Pokémon, con dati quali nome, tipi di attacco, valori di attacco/difesa/eccetera, e se sono o meno leggendari. • Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna; • È verosimile che la prima colonna dovrebbe essere un indice? • Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano? • Se no, settare la prima colonna come indice.

```python
import pandas as pd

df = pd.read_csv("pokemon.csv")

# Dimensioni
print("Dimensioni:", df.shape)

# Anteprima
print(df.head(15))

# Nomi colonne
print(df.columns.tolist())
```

La prima colonna dovrebbe essere un indice?
Nel dataset Pokémon classico (Kaggle), la prima colonna si chiama spesso:# - id - pokedex_number
E contiene valori come:1, 2, 3, 4, 5, ...
Quindi è verosimile che rappresenti l’indice naturale del dataset.

Per verificarlo, basta confrontare: l’indice automatico di Pandas → 0, 1, 2, 3 e la prima colonna → 1, 2, 3, 4, …

```python
#Confronto tra indice Pandas e prima colonna

print(df.index[:10])          # indice Pandas
print(df.iloc[:10, 0])        # prima colonna

#se allora non combaciano:Pandas parte da 0 . Il dataset parte da 1. Questo è un segnale forte che la prima colonna è un vero indice.
```

```python
#Se NON combaciano → dobbiamo sistemare l’indice

df = pd.read_csv("pokemon.csv", index_col=0)

#Verifica finale

print(df.head())
print(df.index)
```

Esercizio 1/2 Dai dataset in piattaforma, carichiamo il file con i dati dei pub, circoli e locali in Italia, chiamato Mappa-dei-pub-circoli-locali-in-Italia.csv Nota: il dataset non ha un encoding UTF-8 ma Latin1; inoltre il separatore non è una virgola, ma un punto e virgola. Dunque per leggerlo dovremo aggiungere i parametri encoding= e sep=, ad esempio: pd.read_csv(file_path, encoding="latin1", sep=";") dove file_path è la posizione del file nel nostro calcolatore.Esercizio 2/2 Esaminiamo il dataset: • quanti dati ci sono in totale? • quali sono i metadati? • stampiamo il primo elemento • stampiamo l'ultimo elemento • riusciamo a stampare un elemento a caso? • quali sono gli anni di inserimento presenti? • quante enoteche ci sono, e come si chiamano? L’ultimo punto potrebbe richiedere molti passaggi, perché i dati in questo dataset non hanno subìto un processo di pulizia e trasformazione adeguato!

```python
#Caricamento del dataset.Il file:non è UTF‑8 → serve encoding="latin1", non usa la virgola → serve sep=";"

df = pd.read_csv(
    "Mappa-dei-pub-circoli-locali-in-Italia.csv",
    encoding="latin1",
    sep=";"
)
```

```python
#analisi del dataset 1) Quanti dati ci sono in totale?

print("Dimensioni del dataset:", df.shape)

#Restituisce (righe, colonne).

#2) Quali sono i metadati?

print("Metadati del dataset:")
print(df.info())

#3) Stampiamo il primo elemento

print("Primo elemento:")
print(df.head(1))

#4) Stampiamo l’ultimo elemento

print("Ultimo elemento:")
print(df.tail(1))

#5) Stampiamo un elemento a caso

print("Elemento casuale:")
print(df.sample(5))


```

```python
#6) Quali sono gli anni di inserimento presenti?
#Prima troviamo la colonna giusta (il dataset è sporco):

print("Nomi colonne:")
print(df.columns.tolist())
#Poi cerchiamo tutte le colonne che contengono “anno”:

print("Colonne che contengono 'anno':")
print([col for col in df.columns if "anno" in col.lower()])
#Supponiamo che la colonna corretta sia "Anno inserimento"

print("Anni di inserimento presenti:")
print(df["Anno inserimento"].unique())
```

```python
#Normalizziamo la colonna Nome

df["nome_norm"] = df["Nome"].str.lower().str.strip()

print("\nEsempi di nome normalizzato:")
print(df["nome_norm"].head())
```

```python
#Filtriamo tutte le enoteche

enoteche = df[df["nome_norm"].str.contains("enoteca", na=False)]

print("\nPrime enoteche trovate:")
print(enoteche.head())
```

```python
#Quante sono?

print("\nNumero totale di enoteche trovate:")
print(len(enoteche))
```

```python
#Come si chiamano?

print("\nNomi delle enoteche:")
print(enoteche["Nome"].unique())
```
